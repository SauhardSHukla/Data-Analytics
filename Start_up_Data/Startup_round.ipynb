##### Problem statement
Given File 'startup_funding.csv'

Find the top 5 startups who received the most number of funding rounds. That means, startups which got fundings maximum number of times.

Print the startup name in descending order with respect to the number of funding round as integer value.

##### Note:
Ola, Flipkart, Oyo, Paytm are important startups, so correct their names. There are many errors in startup names, ignore correcting all, just handle important ones.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path='startup_funding.csv'
file_obj = pd.read_csv(file_path,encoding='utf-8')
file_obj.head(2)
```

```python
# Define a dictionary for mapping old names to new ones
name_mapping = {
    'Oyorooms': 'Oyo',
    'OyoRooms': 'Oyo',
    'Oyo Rooms': 'Oyo',
    'OYO Rooms': 'Oyo',
    'Olacabs': 'Ola',
    'Ola Cabs': 'Ola',
    'Flipkart.com': 'Flipkart',
    'Paytm Marketplace': 'Paytm'
}

# Apply the replacements
file_obj.dropna(subset=['StartupName'], inplace=True)
file_obj.replace({'StartupName':name_mapping}, inplace=True)
data_name=file_obj['StartupName'].value_counts().reset_index(name ='startup_No').sort_values('startup_No',ascending=False)[:5]

#Plotting the Pie chart to show the graphical represntation of the data

plt.figure(figsize=(15,15))
plt.pie(data_name['startup_No'],labels=data_name['StartupName'])
plt.legend(data_name['startup_No'],title ='Round of Funding',loc ='upper right')
```
