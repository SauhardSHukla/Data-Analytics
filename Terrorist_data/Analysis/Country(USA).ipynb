```python
import numpy as np
import csv

# Load the CSV data into NumPy arrays
file_path = 'C:/Users/5530/Downloads/terrorismData.csv'

# Initialize lists to hold the 'Country' and 'Killed' data
countries = []
killed = []

# Open the CSV file and read the data
with open(file_path, mode='r', errors='ignore') as file_obj:
    dict_obj = csv.DictReader(file_obj, skipinitialspace=True)
    
    # Extract 'Country' and 'Killed' columns
    for row in dict_obj:
        countries.append(row['Country'])
        killed.append(row['Killed'])

# Convert the 'countries' and 'killed' lists to NumPy arrays
np_countries = np.array(countries)
np_killed = np.array(killed)

# Replace empty strings in 'np_killed' with '0' (as strings)
np_killed = np.where(np_killed == '', '0', np_killed)

# Convert 'np_killed' to float for numeric operations
np_killed = np_killed.astype(float)

# Create a boolean mask for rows where 'Country' is 'United States'
usa_mask = np_countries == 'United States'

# Extract the 'Killed' values for 'United States' and calculate the sum
usa_killed = np_killed[usa_mask]
total_killed = np.sum(usa_killed)

# Print the total number of killed in the United States
print("Total Killed in United States:", total_killed)
```
